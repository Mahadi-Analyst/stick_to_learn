```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
import scipy
import random
```

```python

# Read the file and specify which column is the date
bakery = pd.read_excel("bakery-.xlsx")

# Output with dates converted to YYYY-MM-DD
bakery["Date"] = pd.to_datetime(bakery["Date"]).dt.strftime("%d-%m-%y")
bakery.to_excel("bakery-_DD_YYYY_MM.xlsx")
bakery
```

```python
#Question: Create a histogram for daily cookie sales

bakery.plot('Date', 'Cookies', kind='hist')
```

```python
# Answer: If the value lies between 0 and 1, then it is symmetrical. So the daily cookie sales is symmetrical. 
bakery.skew()
```

```python
# Question: By the rule of thumb you would expect on 95 percent of all days daily cookie demand will be ____ and ____
```

```python
std = bakery.Cookies.std()
mean = bakery.Cookies.mean()
mean,std
```

```python
# Answer: 95 percent of all days daily cookie demand will be 276 and 805.

(mean-2*std), (mean+2*std)
```
